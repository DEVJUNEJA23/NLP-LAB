```python
import matplotlib.pyplot as plt
import random
import nltk
from nltk.corpus import stopwords

# Sample text data
text_data = """
This is a sample text for creating a word cloud. Word clouds are a fun way to visualize text data.
You can customize the appearance by adjusting the font size, color, and layout.
In this example, we'll create a basic word cloud without using external libraries like wordcloud or Counter.text_data = text_data.lower()"""

words = text_data.split()

stopwords = set(stopwords.words('english'))
words = [word for word in words if word not in stopwords]


word_freq = {}


for word in words: 
    if word.isalpha():  
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

plt.figure(figsize=(10, 6))
grid_size = 20
grid = [[False] * grid_size for _ in range(grid_size)]

for word, freq in sorted_word_freq:
    placed = False
    while not placed:
        x = random.randint(0, grid_size - 1)
        y = random.randint(0, grid_size - 1)

        if not grid[x][y]:
            grid[x][y] = True
            
            placed = True

    color = (random.random(), random.random(), random.random())  # Random RGB color
    plt.text(x / grid_size, y / grid_size, word, fontsize=freq * 15, ha='center', va='center', color=color)

plt.axis("off")
plt.show()
```
